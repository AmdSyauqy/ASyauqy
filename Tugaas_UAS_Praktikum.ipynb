```python
class Kursus:
    def __init__(self, nama_kursus, kode_kursus, durasi, kapasitas):
        """
        Konstruktor untuk membuat objek Kursus
        """
        try:
            self.nama_kursus = nama_kursus
            self.kode_kursus = kode_kursus
            self.durasi = int(durasi)
            self.kapasitas = int(kapasitas)
            self.jumlah_peserta = 0
        except ValueError:
            raise ValueError("Durasi dan kapasitas harus berupa angka yang valid.")

    def tambah_peserta(self):
        """
        Menambah peserta ke kursus jika kapasitas belum penuh.
        """
        if self.jumlah_peserta < self.kapasitas:
            self.jumlah_peserta += 1
            return f"Peserta berhasil ditambahkan. Jumlah peserta saat ini: {self.jumlah_peserta}."
        else:
            raise Exception("Kapasitas kursus sudah penuh.")

    def info_kursus(self):
        """
        Mengembalikan informasi lengkap tentang kursus
        """
        return (f"Nama Kursus: {self.nama_kursus}, "
                f"Kode: {self.kode_kursus}, "
                f"Durasi: {self.durasi} minggu, "
                f"Kapasitas: {self.kapasitas}, "
                f"Jumlah Peserta: {self.jumlah_peserta}")


daftar_kursus = []
kursus_dict = {}

def tampilkan_semua_kursus():
    """
    Fungsi untuk menampilkan semua kursus yang tersedia
    """
    if not daftar_kursus:
        return "Belum ada kursus yang tersedia."
    return "\n".join(kursus.info_kursus() for kursus in daftar_kursus)

def cari_kursus(kode):
    """
    Fungsi untuk mencari kursus berdasarkan kode
    """
    try:
        kursus = kursus_dict[kode]
        return kursus.info_kursus()
    except KeyError:
        return "Kursus dengan kode tersebut tidak ditemukan."

def tambah_kursus():
    """
    Fungsi untuk menambahkan kursus baru
    """
    nama = input("Masukkan nama kursus: ")
    kode = input("Masukkan kode kursus: ")
    durasi = input("Masukkan durasi kursus (dalam minggu): ")
    kapasitas = input("Masukkan kapasitas kursus: ")

    if kode in kursus_dict:
        return "Kursus dengan kode ini sudah ada."

    try:
        kursus_baru = Kursus(nama, kode, durasi, kapasitas)
        daftar_kursus.append(kursus_baru)
        kursus_dict[kode] = kursus_baru
        return "Kursus berhasil ditambahkan."
    except ValueError as e:
        return str(e)

def tambah_peserta_ke_kursus():
    """
    Fungsi untuk menambah peserta ke kursus berdasarkan kode
    """
    kode = input("Masukkan kode kursus untuk menambah peserta: ")

    try:
        kursus = kursus_dict[kode]
        return kursus.tambah_peserta()
    except KeyError:
        return "Kursus dengan kode tersebut tidak ditemukan."
    except Exception as e:
        return str(e)

if __name__ == "__main__":
    while True:
        print("\nMenu:")
        print("1. Tambah Kursus")
        print("2. Tambah Peserta ke Kursus")
        print("3. Cari Kursus Berdasarkan Kode")
        print("4. Tampilkan Semua Kursus")
        print("5. Keluar")

        pilihan = input("Pilih menu (1/2/3/4/5): ")

        if pilihan == '1':
            print(tambah_kursus())
        elif pilihan == '2':
            print(tambah_peserta_ke_kursus())
        elif pilihan == '3':
            kode = input("Masukkan kode kursus yang ingin dicari: ")
            print(cari_kursus(kode))
        elif pilihan == '4':
            print(tampilkan_semua_kursus())
        elif pilihan == '5':
            print("Terima kasih, program selesai.")
            break
        else:
            print("Pilihan tidak valid, coba lagi.")

```
